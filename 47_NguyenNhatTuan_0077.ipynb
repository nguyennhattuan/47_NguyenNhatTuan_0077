A Hướng Đối Tượng

```python
class NhanVien:
    def __init__(self, ma_nv: str, luong_cb: float):
        self.__ma_nv = ma_nv
        self.__luong_cb = luong_cb

    @property
    def ma_nv(self) -> str:
        return self.__ma_nv

    @property
    def luong_cb(self) -> float:
        return self.__luong_cb

    def tinh_luong(self) -> float:
        # Mặc định trả về lương cơ bản
        return self.__luong_cb


class NhanVienTangCa(NhanVien):
    def tinh_luong(self) -> float:
        return self.luong_cb * 1.2


class NhanVienChuyenChuyen(NhanVien):
    def tinh_luong(self) -> float:
        return self.luong_cb + 500_000


# Phần main minh họa yêu cầu
if __name__ == "__main__":
    danh_sach = [
        NhanVien("NV001", 10_000_000),
        NhanVienTangCa("NV002", 10_000_000),
        NhanVienChuyenChuyen("NV003", 10_000_000),
    ]

    for nv in danh_sach:
        print(f"Ma NV: {nv.ma_nv} - Luong: {nv.tinh_luong():,.0f} VND")

```

B Numpy

```python
import numpy as np

# Dữ liệu ban đầu
san_luong = [12, None, 8, 15, None, 10]

# 1. Chuyển thành mảng NumPy (ép kiểu float để xử lý NaN)
arr = np.array(san_luong, dtype=float)

# Reshape thành mảng 2x3
arr_reshaped = arr.reshape(2, 3)
print("Mảng sau khi reshape (2x3):")
print(arr_reshaped)
print()

# 2. Tính giá trị trung bình (bỏ qua NaN)
mean_value = np.nanmean(arr_reshaped)
print(f"Giá trị trung bình (bỏ qua NaN): {mean_value}")
print()

# Thay thế NaN bằng giá trị trung bình
arr_clean = arr_reshaped.copy()           
arr_clean[np.isnan(arr_clean)] = mean_value

print("Mảng sau khi thay NaN bằng trung bình:")
print(arr_clean)
print()

# 3. Trích xuất các phần tử > 9
greater_than_9 = arr_clean[arr_clean > 9]

print("Các phần tử > 9:", greater_than_9)

# Tìm giá trị nhỏ nhất trong số các phần tử > 9
if len(greater_than_9) > 0:
    min_value = np.min(greater_than_9)
    print(f"Giá trị nhỏ nhất trong các phần tử > 9: {min_value}")
else:
    print("Không có phần tử nào > 9")
```

Lọc dữ liệu

```python
import pandas as pd
import os

# Đường dẫn thư mục
path = r'D:\47_NguyenNhatTuan_0077\Data'

# 1. Đọc dữ liệu trực tiếp từ file csv trong thư mục
file_csv = os.path.join(path, 'nhan_su.csv')
df = pd.read_csv(file_csv)
print(df.head())
print(df.info())
```

```python
# Lọc các bản ghi có Ma_NV chứa chuỗi "NS"

mask_ns = df['Ma_NV'].str.contains('NS', case=False, na=False)

# Nam_Cong_Tac >= 2021
mask_year = df['Nam_Cong_Tac'] >= 2021

df_filtered = df[mask_ns & mask_year]

print(df_filtered)
```

Làm sạch dữ liệu

```python

df_filtered = df_filtered.copy()

# Tính giá trị trung bình của cột Thanh_Tich
mean_thanh_tich = df_filtered['Thanh_Tich'].mean()

# Điền giá trị thiếu bằng giá trị trung bình
df_filtered['Thanh_Tich'] = df_filtered['Thanh_Tich'].fillna(mean_thanh_tich)

print(f"Mean value used for filling: {mean_thanh_tich}")
print(df_filtered)
```

Tổng hợp dữ liệu (Groupby)

```python
df_cleaned = df_filtered

#  Tinh tong thanh tich theo phong ban
grouped_data = df_cleaned.groupby('Phong_Ban')['Thanh_Tich'].sum().reset_index()

print(grouped_data)
```

Biến đổi bảng (Pivot Table)

```python
# Read the cleaned data
df_cleaned = pd.read_csv('nhan_su_cleaned.csv')


# Pivot Table
pivot_table = df_cleaned.pivot_table(
    index='Phong_Ban',
    columns='Nam_Cong_Tac',
    values='Thanh_Tich',
    aggfunc='sum',
    fill_value=0
)

print(pivot_table)
```

Lưu trữ và Truy vấn

```python
import sqlite3

conn = sqlite3.connect('HeThong.db')

pivot_table.reset_index().to_sql('bang_luong_2024', conn, if_exists='replace', index=False)

# Thực hiện truy vấn hiển thị 5 dòng đầu tiên của bảng
query = "SELECT * FROM bang_luong_2024 LIMIT 5"
df_check = pd.read_sql(query, conn)

# Đóng kết nối
conn.close()

print(df_check)
```
